```python
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel('Dataset.xlsx', sheet_name="E Comm")
```

```python
print(data)
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

#fill missing values with o


```python
columns_to_fill = ['Tenure', 'WarehouseToHome', 'HourSpendOnApp', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder'] 
data[columns_to_fill] = data[columns_to_fill].fillna(0)    
print(data)  
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```

handling outliers

```python
import seaborn as sns
columns_to_remove_outliers = ['Tenure', 'WarehouseToHome', 'HourSpendOnApp', 'OrderAmountHikeFromlastYear', 'CouponUsed', 'OrderCount', 'DaySinceLastOrder']
sns.boxplot(data[columns_to_remove_outliers])
```

```python
Q1 = data['WarehouseToHome'].quantile(0.25)
Q3 = data['WarehouseToHome'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data = data[(data['WarehouseToHome'] >= lower_bound) & (data['WarehouseToHome'] <= upper_bound)]
```
